```python
from pylab import *
%matplotlib inline
```

```python
def f(x,omega=1):
    return sin(omega*x)

def g(h_x,x,args=[]):
    if len(args)==0:
        resultado = exp(-x)*h_x(x)
    else:
        resultado = exp(-x)*h_x(x,args)
    return resultado
```

```python
x=linspace(0,2*pi,200)
for i in range(5):
    plot(x,f(x,i))
    plot(x,g(f,x,[i]))
show()
```
